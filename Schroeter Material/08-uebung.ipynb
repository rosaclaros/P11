Verändern Sie den Code aus Übungsblatt zum Thema 3, Aufgabe 1: "Taschenrechner mit Auswahl der Grundrechenarten und Speicherung des Ergebnisses" so, dass mögliche Ausnahmen nun durch `try  - except` Anweisungen abgefangen werden.

**Tipp**: Wenn Sie unsicher sind, welche Ausnahmen sie abfangen sollen, testen Sie ihr Programm mit fehlerhaften Eingaben und fangen Sie die den auftretenden Fehlermeldungen entsprechenden Fehler an geeigneter Stelle ab. Beachten Sie dabei auch die *Exception-Hierarchie* von Python (https://docs.python.org/3/library/exceptions.html).

Berücksichtigen Sie auf jeden Fall folgende Ausnahmen:
- `ZeroDivisionError` bei Division durch 0
- `ValueError` bei Eingabe eines Strings statt Zahlen
- `ValueError` bei Eingabe eines Wertes außerhalb des `rechenart`-Wertebereichs (diese Exception muss mit `raise` definiert werden)


```python
## ANGABE Aufgabe 2a (=Code aus Musterlösung Aufgabe 1, Blatt4):


zahl_1 = ""
zahl_2 = ""

print("Herzlich willkommen zu meinem Taschenrechner!\n")


##INPUT ZAHLEN:
zahl_1 = float(input("Bitte geben Sie die erste Zahl ein >> "))
zahl_2 = float(input("Bitte geben Sie die zweite Zahl ein >> "))
   
    
##INPUT RECHENART:    
rechenart = input("Welche Rechenart soll angewendet werden? ... Wählen Sie:"
                    "\n1 für Addition"
                    "\n2 für Subtraktion"
                    "\n3 für Multiplikation"
                    "\n4 für Division"
                    "\n5 für Modulo\n >>")

rechenart = int(rechenart)

##VERARBEITUNG RECHENART:
if 0 < rechenart < 6:

##BERECHNUNG ERGEBNIS:    
    if rechenart == 1: #Addition
        rechenzeichen = " + "
        rechenergebnis = zahl_1 + zahl_2
        
    if rechenart == 2: #Subtraktion
        rechenzeichen = " - "
        rechenergebnis = zahl_1 - zahl_2
        
    if rechenart == 3: #Multiplikation
        rechenzeichen = " x "
        rechenergebnis = zahl_1 * zahl_2
        
    if rechenart == 4: #Division
        rechenzeichen = " : "
        rechenergebnis = zahl_1 / zahl_2
        
    if rechenart == 5: #Modulo
        rechenzeichen = " // "
        rechenergebnis = zahl_1 // zahl_2
    
    ergebnis = str(zahl_1) + rechenzeichen + str(zahl_2) + " = " + str(rechenergebnis)
    
    
    print("\nErgebnis:", ergebnis)
    
else: print("Für den eingegebenen Wert ist keine Rechenart definiert.")
```
